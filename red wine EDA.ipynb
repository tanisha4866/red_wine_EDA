```python
# # Q2. How did you handle missing data in the wine quality data set during the feature engineering process?
# # Discuss the advantages and disadvantages of different imputation techniques.

# Missing values can be handled using either Mean,Median or mode. It depends on the type of data if it contains quliers or not.
# for example if a dataset contain big numbered oultliers then Median will be used to fill the missing values and in the vice 
# versa condition mean will be used. Mode will be used when the dataset has less variety of value so the most popular values 
# will fill in.

# Mean/Median Imputation:

# Advantages:
# Simple and quick to implement.
# Works well when the missing data is missing completely at random (MCAR) or missing at random (MAR).

# Disadvantages:
# Doesn't capture the underlying distribution or relationships in the data.
# Can introduce bias and underestimate variability if the missingness is not MCAR or MAR.
# Distorts the original variable's distribution if the missingness is substantial.


# Mode Imputation:

# Advantages:
# Suitable for categorical variables with non-ordinal or non-numeric values.
# Preserves the mode of the original distribution.

# Disadvantages:
# May not work well if the missingness is related to unobserved factors that affect the mode.
```
